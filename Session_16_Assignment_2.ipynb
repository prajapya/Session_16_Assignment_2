```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from sklearn import decomposition
from sklearn.decomposition import PCA
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
print("First 5 samples:")
iris.data[0:5]
```

```python
A0 = iris.data
```

```python
print("Dimensions:", A0.shape)
```

```python
print("Feature Names:", iris.feature_names)
```

```python
#Using sklearn.decomposition.PCA

pca = PCA(n_components=3)
pca.fit(A0)
```

```python
print("Principal components:")
print(pca.components_)
```

```python
print("4D to 3D compression:")
X = pca.transform(A0)
print(pca.transform(A0)[:5,:])
```

```python
X.shape
```

```python
g1 = (X[0:50,0], X[0:50,1], X[0:50,2])
g2 = (X[50:100,0], X[50:100,1], X[50:100,2])
g3 = (X[100:150,0], X[100:150,1], X[100:150,2])
data = (g1, g2, g3)
colors = ('red', 'green', 'blue')
groups = ('g1', 'g2', 'g3')

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, facecolor="1.0")
ax = fig.gca(projection='3d')

for dt, color, group in zip(data, colors, groups):
    x, y, z = dt
    ax.scatter(x, y, z, alpha=0.8, c=color, edgecolors='none', s=30, label=group)
plt.title("Iris 3D PCA plot")
plt.legend(loc=2)
plt.show()
```

<span style="color:black; font-size:1.1em;">**Thus using PCA we have transformed Iris 4D data into 3D and plotted in 3D graph using Axes3D**</span>
